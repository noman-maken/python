```python
import math
option_type=input('''Calculate perimeter, area, and volume of different geometrical shapes\n
Enter p for perimeter\nEnter a for area\nEnter v for volume\n''')
if option_type=="p" or option_type=="a" or option_type=="v":
    option_shape=input('''Enter geometrical shape: \nEnter s for square\nEnter r for rectangle
Enter t for triangle\nEnter c for circle.\n''')
    if option_shape=="s" or option_shape=="r" or option_shape=="t" or option_shape=="c":
        if option_shape=="s":
            length_of_side=int(input("Enter the length of the side of square :"))
            if option_type=="p":
                perimeter_of_square=4*length_of_side
                print("Perimeter of your square is : ",perimeter_of_square )
            elif option_type=="a":
                area_of_square=length_of_side*length_of_side
                print("Area of your square is: ",area_of_square)
            elif option_type=="v":
                volume_of_square=length_of_side*length_of_side*length_of_side
                print("Volume of cube is:  ",volume_of_square)
        elif option_shape=="r":
            length_of_rectangle=int(input("Enter the length of rectangle: "))
            width_of_rectangle=int(input("Enter the width of rectangle: "))
            if option_type=="p":
                perimeter_of_rectangle=((2*length_of_rectangle)+(2*width_of_rectangle))
                print("Perimeter of your rectangle is : ",perimeter_of_rectangle )
            elif option_type=="a":
                area_of_rectangle=length_of_rectangle*width_of_rectangle
                print("Area of rectangle is: ",area_of_rectangle)
            elif option_type=="v":
                height_of_rectangle=int(input("Enter the height of rectangular prism: "))
                volume_of_rectangle=length_of_rectangle*width_of_rectangle*height_of_rectangle
                print("Volume of your rectangular prism is: ",volume_of_rectangle)
        elif option_shape=="c":
            radius_of_circle=int(input("Enter a radius of your circle: "))
            if option_type=="p":
                perimeter_of_circle=2*3.14*radius_of_circle
                print("Perimeter of circle is: ",perimeter_of_circle)
            elif option_type=="a":
                area_of_circle=3.14*radius_of_circle*radius_of_circle
                print("Area of circle is: ",area_of_circle)
            elif option_type=="v":
                volume_of_circle=(4/3)*3.14*radius_of_circle*radius_of_circle*radius_of_circle
                print("Volume of sphere is:  ",volume_of_circle)
        elif option_shape=="t":
            first_side_of_triangle=int(input("Enter the length of first side of triangle: "))
            second_side_of_triangle=int(input("Enter the length of second side of triangle: "))
            third_side_of_triangle=int(input("Enter the length of third side of triangle: "))
            if option_type=="p":
                perimeter_of_triangle=first_side_of_triangle+second_side_of_triangle+third_side_of_triangle
                print("Perimeter of your triangle is : ",perimeter_of_triangle)
            elif option_type=="a":
                perimeter_of_triangle=first_side_of_triangle+second_side_of_triangle+third_side_of_triangle
                semiperimeter_of_triangle=perimeter_of_triangle/2
                calculation=(semiperimeter_of_triangle*(semiperimeter_of_triangle-first_side_of_triangle)(semiperimeter_of_triangle-second_side_of_triangle)(semiperimeter_of_triangle-third_side_of_triangle))
                area_of_triangle=math.sqrt(calculation)
                print("Area of triangle is : ",area_of_triangle)
            elif option_type=="v":
                height_of_triangle=int(input("Enter the height of triangular prism: "))
                volume_of_triangle=height_of_triangle*area_of_triangle
                print("Volume of triangular prism is : ",volume_of_triangle)
    else:
        print("Invalid Input ")
else:
    print("Invalid Input ")
```
